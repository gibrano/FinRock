```python
client = OpenAI()
```

```python
NEWS_AGENT_SYSTEM_PROMPT = """
You are a News Intelligence Agent.

Your goal is to fetch and analyze news articles from URLs.

TASKS:
1. Retrieve the webpage and extract only article text.
2. Extract the news content.
3. Summarize the content clearly.

OUTPUT FORMAT (JSON):
{
  "headline": "",
  "url": "",
  "content": "",
  "summary": ""
}
"""
```

```python
def analyze_news_url(url):
    response = client.responses.create(
        model="gpt-5",
        input=[
            {"role": "system", "content": NEWS_AGENT_SYSTEM_PROMPT},
            {"role": "user", "content": f"Analyze this news article: {url}"}
        ],
        tools=[{"type": "web_search"}],
    )

    return response #.choices[0].message
```
